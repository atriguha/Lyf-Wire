## Variables

Each patient is represented in the data set by six biomechanical attributes derived from the shape and orientation of the pelvis and lumbar spine.

**Specifically**:
#### Pelvic Incidence
    Angle between the line perpendicular to the sacral plate at its midpoint and the line connecting this point to the femoral heads axis. Pelvic incidence can vary from 33° to 85° in the normal population.
#### Pelvic Tilt
    Angle created by a line running from the sacral endplate midpoint to the center of the bifemoral heads and the vertical axis. The average ranges of anterior and posterior pelvic tilting are 13.0 ± 4.9°, and 8.9 ± 4.5°, respectively.
#### Lumbar Lordosis Angle
    Angle between the upper plane of the L1 lumbar vertebrae and the upper plane of the S1 sacral vertebrae; normal lordosis (NL) is in the 31–50° range
#### Sacral Slope
    Angle of the sacral plateau to the horizontal. The normal range of value for the sacral slope was from 32° to 49°
#### Pelvic Radius
    The distance from the hip axis to the posterior-superior corner of the S1 endplate.
#### Grade of Spondylolisthesis
    Spondylolisthesis is a spinal disorder in which a bone (vertebra) slips forward onto the bone below it. The seriousness of the disorder is marked by grade values

# Data Pre-Processing

```python
# Let's convert these categorical values into numeric
weka_3C["class"].replace(["Hernia", "Spondylolisthesis", "Normal"], [0, 1, 2], inplace=True)
```

```python
# split dataset into x and y
y = weka_3C["class"].values # classes
x_data = weka_3C.drop(["class"], axis=1) # features
```

```python
# train-test-split time - 15% for test, 85% for train 
x_train, x_test, y_train, y_test = train_test_split(x_data, y, test_size=0.15, random_state=42)
```

```python
# normalization
scaler = MinMaxScaler()
scaler.fit_transform(x_train)
scaler.transform(x_test)
```

<a id=2></a>
# Selected Algorithms

K-Nearest Neighbor, Support Vector Machines, Random Forests, and Decision Tree Algorithms are used for this project.


```python
# KNN algorithm

# looking for best k value
score_list = []
for each in range(1,50):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))

plt.plot(range(1,50),score_list)
plt.xlabel("k values")
plt.ylabel("Accuracy")
plt.title("Finding the Best k Value")
plt.show()
```

According to this graphic, it seems that the best k value can be 2 or 14. We shall use both


```python
neighbor = 14
knn = KNeighborsClassifier(n_neighbors = neighbor)
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
print(" {} KNN Model Test Accuracy: {} ". format(neighbor, knn.score(x_test, y_test)))
```

```python
neighbor = 2
knn = KNeighborsClassifier(n_neighbors = neighbor)
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
print(" {} KNN Model Test Accuracy: {} ". format(neighbor, knn.score(x_test, y_test)))
```

Selecting the k value as 31 results better than selecting the k value as 32.


```python
# Support Vector Machine Algorithm

# train
svm = SVC(random_state=1)
svm.fit(x_train,y_train)

# test
print("SVM Model Accuracy: {}".format(svm.score(x_test,y_test)))
```

Interestingly, SVM accuracy is exactly equal to the KNN algorithm with k=2 or 14.

```python
# Decision Tree Algorithm

# train
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

# test
print("DT Model Accuracy: {}".format(dt.score(x_test,y_test)))
```

```python
# Random Forests Algorithm

# train
rf = RandomForestClassifier(n_estimators = 100, random_state = 1) # for 100 trees
rf.fit(x_train.values,y_train)
# test
print("RF Model Accuracy: {}".format(rf.score(x_test.values,y_test)))
```

Indeed the SVM and KNN are best performing algorithms here, However we can perform ensemble learning for robustness.

# Ensemble Learning

```python
from sklearn.ensemble import VotingClassifier

#create a dictionary of our models
estimators = [('knn', knn), ('rf', rf), ('svm', svm)]

#create our voting classifier, inputting our models
ensemble = VotingClassifier(estimators, voting='hard')
ensemble.fit(x_train.values,y_train)
# test
print("Ensemble Model Accuracy: {}".format(ensemble.score(x_test.values,y_test)))
```
